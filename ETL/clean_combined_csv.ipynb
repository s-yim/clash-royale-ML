```python
import os
import pandas as pd

# File path to the CSV file
csv_file_path = os.path.join('..', 'Resources', 'combined_player_tags1.csv')

# Columns to drop
columns_to_drop = ['/player_tag', '/data/currentPathOfLegendSeasonResult', 
                   '/data/lastPathOfLegendSeasonResult', '/data/lastPathOfLegendSeasonResult/rank']

# Load the CSV file into a DataFrame
player_data_wide_df = pd.read_csv(csv_file_path)

# Dropping columns
player_data_wide_df.drop(columns=columns_to_drop, inplace=True)

# Displaying the DataFrame
player_data_wide_df.head()
```

```python
player_data_wide_df.columns
```

```python
url_columns_to_drop = [col for col in player_data_wide_df.columns if 'Urls' in col]
player_data_wide_df.drop(columns=url_columns_to_drop, inplace=True)
player_data_wide_df.head()
```

```python
info_columns_to_drop = [col for col in player_data_wide_df.columns if 'completionInfo' in col]
player_data_wide_df.drop(columns=info_columns_to_drop, inplace=True)
player_data_wide_df.head()
```

```python
player_data_wide_df.to_csv('combined_player_tags_master.csv', index=False)
```
