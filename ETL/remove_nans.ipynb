```python
import os
import pandas as pd
import numpy as np

# File path to the CSV file
csv_file_path = os.path.join('..', 'Resources', 'combined_player_tags_master.csv')

# Load the CSV file into a DataFrame
player_data_df = pd.read_csv(csv_file_path)
```

```python
player_data_df.head()
```

```python
# Find columns with NaNs and separate columns by integer and string
columns_with_nans = player_data_df.columns[player_data_df.isna().any()].tolist()
int_columns_with_nans = player_data_df[columns_with_nans].select_dtypes(include=np.number).columns.tolist()
string_columns_with_nans = player_data_df[columns_with_nans].select_dtypes(include='object').columns.tolist()
```

```python
# Replace NaNs in columns with 0 for integer columns and "NO" for string columns
int_columns_to_replace = int_columns_with_nans
string_columns_to_replace = string_columns_with_nans  
player_data_df[int_columns_to_replace] = player_data_df[int_columns_to_replace].fillna(0)
player_data_df[string_columns_to_replace] = player_data_df[string_columns_to_replace].fillna("NO")
```

```python
# Overwrite file 
player_data_df.to_csv('combined_player_tags_master.csv', index=False)
```
