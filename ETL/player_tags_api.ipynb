```python
import requests
import random
import json
from datetime import datetime
from config import clash_api_key

# Clash Royale API endpoint
api_endpoint = "https://api.clashroyale.com/v1/players/{}"

# List to store player tags and player data
player_data_list = []

# Function to make API request and process the response
def make_api_request(player_tag):
    api_url = api_endpoint.format(player_tag)
    headers = {
        "Authorization": "Bearer %s" % clash_api_key, 
    }
    response = requests.get(api_url, headers=headers)

    # Process the API response
    if response.status_code == 200:
        # If API call is successful, add player data to the list
        player_data = response.json()
        player_data_list.append({"player_tag": player_tag, "data": player_data})
    else:
        print(f"Failed to fetch data for player tag {player_tag}. Status code: {response.status_code}")

# Function to save JSON data to a local file with timestamp
def save_json_data(data, file_prefix='player_tags'):
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    file_name = f"{file_prefix}_{timestamp}.json"
    with open(file_name, 'w') as file:
        json.dump(data, file, indent=2)
    return file_name

# Loop 10 times
for iteration in range(10):
    # List to store player tags and player data for each iteration
    player_data_list = []

    # Generate 2000 random player tags
    for _ in range(2000):
        tag_length = random.randint(5, 9)
        player_tag = ''.join(random.choices('PYLQRJCUV0289', k=tag_length))
        formatted_player_tag = "%23" + player_tag  
        data = make_api_request(formatted_player_tag)
        if data:
            player_data_list.append(data)

    # Save as JSON with timestamp
    output_file = save_json_data(player_data_list, file_prefix=f'player_tags_{iteration}')
    print(f"Iteration {iteration + 1}: JSON data saved to: {output_file}")
```
